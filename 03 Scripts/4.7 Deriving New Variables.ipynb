#03. Exercise 4.7

```python
# Create a subset
df = df_orders_products_merged[:1000000]
```

```python
# Check the shape
df.shape
```

```python
# Define a function
def price_label(row):
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check frequency of products in subset
df['price_range'].value_counts(dropna = False)
```

```python
# Find maximum price in subset
df['prices'].max()
```

```python
# Apply value for high range product in subset
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
# Apply value for mid range product in subset
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
# Apply value for low range product in subset
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
# Check frequency of products in subset
df['price_range_loc'].value_counts(dropna = False)
```
